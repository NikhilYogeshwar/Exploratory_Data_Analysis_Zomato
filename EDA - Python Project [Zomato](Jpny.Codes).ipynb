```python
final_df.Country.value_counts().index                # Here we got all the countries name in the list.
```

```python
country_names = final_df.Country.value_counts().index               # We saved this list in 'country_name'.
```

```python
 final_df.Country.value_counts().values                # Here we are getting the values of counts acc. to countries.
```

```python
country_val = final_df.Country.value_counts().values    # Here now we named that list to this.
```

```python
# Pie Chart
plt.pie(country_val,labels=country_names)             # Here we are plotting an pie-chart (x-axis values,y-axis values).
```

```python
# Pie Chart - Top 3 countries that uses Zomato
plt.pie(country_val[:3],labels=country_names[:3])            # Here we are plotting an pie-chart for top 3 (x-axis values,y-axis values).
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')   # Here we are adding percentage to the chart.
```

Observation: Zomato's maximum records or transactions are from India, then USA and then UK.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])      # Here we have grouped these 3 columns.
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()   # We can see the columns with values also.
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()       # Here we reseted the index.
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
                                                # Here we have renames that 0 column to rating count.
```

```python
ratings =final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
                            # Here we named that table to 'ratings' now.
```

```python
ratings
```

## Observation
1. When rating is btw 4.5 to 4.9----> Excellent
2. When rating is btw 4.0 to 4.4----> Very good
3. When rating is btw 3.5 to 3.9----> Good
4. When rating is btw 3.0 to 3.4----> Averege
5. When rating is btw 2.5 to 2.9---->Average
6. When rating is btw 2.0 to 2.4---->Poor
